```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Summer.csv')
df.head()
```

# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
data=[]
for City in df['City'].unique():
    data.append([City,len(df[df['City']==City])])
pd.DataFrame(data,columns=['City','freq']).sort_values(by='freq',ascending=False).plot(x='City',y='freq',kind='bar',figsize=(10,5))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
goldM = df[(df.Medal == 'Gold')]
tg=goldM.Sport.value_counts().reset_index(name='Medal').head()
tg
```

```python
g = sns.catplot(x="index", y="Medal", data=tg,height=6, kind="bar", palette="muted")
g.despine(left=True).set_xlabels("Top 5 sports").set_ylabels("Number of gold Medals")

plt.title('Gold medals per Sports')
```

```python
print('Aquatics is having most number of Gold Medals so far.')
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
tm=df.Sport.value_counts().reset_index(name='Medal').head()
tm
```

```python
g = sns.catplot(x="index", y="Medal", data=tm,height=6, kind="bar", palette="muted")
g.despine(left=True).set_xlabels("Top 5 sports").set_ylabels("Number of Medals")
g.set_xlabels("Top 5 sports")
g.set_ylabels("Number of Medals")
plt.title('Medals per Sports')
```

```python
print('Aquatics is having most number of Medals so far.')
```

### 4. Which player has won most number of medals? (Top 5)

```python
ta=df.Athlete.value_counts().reset_index(name='Medal').head()
ta
```

```python
g = sns.catplot(x="index", y="Medal", data=ta,height=6, kind="bar", palette="muted")
g.despine(left=True).set_xlabels("Top 5 players").set_ylabels("Number of Medals")

plt.title('Medals per Athlete')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
goldM = df[(df.Medal == 'Gold')]
goldM = df[(df.Country == 'IND')]
goldM.head()
```

```python
print('India won first gold medal in summer olympics in',goldM.Year.min())
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data=[]
for Event in df['Event'].unique():
    data.append([Event,len(df[df['Event']==Event])])
pd.DataFrame(data,columns=['Event','freq']).sort_values(by='freq',ascending=False).plot(x='Event',y='freq',kind='bar',figsize=(100,50))
```

```python
print('Football is the most popular event')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
goldM = df[(df.Medal == 'Gold')]
goldM = df[(df.Gender == 'Women')]
goldM.head()
tgf=goldM.Sport.value_counts().reset_index(name='Medal').head()
tgf
```

```python
g = sns.catplot(x="index", y="Medal", data=tgf,height=6, kind="bar", palette="muted")
g.despine(left=True).set_xlabels("Top 5 female players").set_ylabels("Number of Gold medals")

plt.title('Gold medals per female Athlete')
```

```python
print('Aquatics sport is having most female Gold Medalists')
```
